```python
import sagemaker
from sagemaker.pytorch import PyTorch
```

```python
sagemaker_session = sagemaker.Session()

bucket = 'slip-ml'

role = 'arn:aws:iam::438465160412:role/Sagemaker'
```

```python
sagemaker.image_uris.retrieve(framework='pytorch',
                             region=sagemaker_session.boto_region_name,
                             instance_type='ml.p3.2xlarge',
                             image_scope='training')
```

```python
estimator = PyTorch(
    entry_point="lipnet.py",
    role=role,
    image_uri='763104351884.dkr.ecr.us-east-1.amazonaws.com/pytorch-training:2.6.0-gpu-py312',
    instance_count=1,
    instance_type="ml.p3.2xlarge",
    source_dir="source",
    output_path=f's3://{bucket}/models',
    code_location=f's3://{bucket}/model-building'
)
```

```python
estimator.fit({'training': f's3://{bucket}/data/preprocessing-2/train',
                'test': f's3://{bucket}/data/preprocessing-2/test'})
```
